```python
# 18. Count how many matches each umpire officiated

u1 = df['Umpire1'].value_counts()
u2 = df['Umpire2'].value_counts()

u1_u2 = u1.add(u2,fill_value=0).sort_values(ascending=False).astype(int).reset_index()
u1_u2.columns = ['Umpire', 'Count']
u1_u2
```

```python
# 19. Find matches where both umpires were the same pair more than once
df['u1+u2'] = df['Umpire1'] + df['Umpire2']

# get counts
umpire_counts = df['u1+u2'].value_counts()

# assign count back as new column
df['UmpireCount'] = df['u1+u2'].map(umpire_counts)

# filter df for more than 1 count
df[df['UmpireCount'] > 1]
```

```python
# 21. Find matches where the Player of the Match was from the losing team

# First create loosing team column
loosing_team_list = []
for i in range(len(df)):
    if df.loc[i,'Team1'] != df.loc[i, 'WinningTeam']:
        loosing_team_list.append('Team1')
    else:
        loosing_team_list.append('Team2')

# loosing team column
df.loc[:, 'Loosing_Team'] = loosing_team_list

# second filter df
pom_loo_team = []  # List of Player of the match in loosing team
for i in range(len(df)):
    pom = df.loc[i, 'Player_of_Match']
    if pd.isna(pom):  # skip if no player of the match
        continue
    if (df.loc[i,'Loosing_Team'] == 'Team1') and (pom in df.loc[i,'Team1Players']):
        pom_loo_team.append(int(df.loc[i,'ID']))
    elif (df.loc[i,'Loosing_Team'] == 'Team2') and (pom in df.loc[i,'Team2Players']):
        pom_loo_team.append(int(df.loc[i,'ID']))

df_filtered = df[df['ID'].isin(pom_loo_team)]
df_filtered
```

```python
# 23. Write a function that returns the win-loss ratio for a team in a given season

def win_loss(year, team):
    a = df[df['Season'] == year]
    t1 = a['Team1'].value_counts()
    t2 = a['Team2'].value_counts()
    
    total = t1.add(t2, fill_value=0)
    won = a['WinningTeam'].value_counts()
    
    t = total.get(team, 0)
    w = won.get(team, 0)
    l = t - w
    
    return w / l if l > 0 else None

a = win_loss('2012','Pune Warriors')
float(a)
```

```python
# 24. Write a function that returns the head-to-head record of two teams in a given city

def head_to_head(city,t1,t2):
    city = df[df['City'] == city]
    
    record = city[((city['Team1'] == t1) | (city['Team2'] == t1)) &
            ((city['Team1'] == t2) | (city['Team2'] == t2))]

    return record

head_to_head('Mumbai','Mumbai Indians','Chennai Super Kings')
```

```python
# 25. Write a function that returns the track record of two teams against each other (overall)

def track_rec(t1,t2):
    two_teams = df[((df['Team1'] == t1) | (df['Team2'] == t1)) &
                    ((df['Team1'] == t2) | (df['Team2'] == t2))]
    
    return two_teams

track_rec('Pune Warriors','Chennai Super Kings')
```

```python
# 28. Find the most frequent Player of the Match against a specific team.

def most_frequent(team_name):
    team = df[(df['Team1'] == team_name) | (df['Team2'] == team_name)]
    return team['Player_of_Match'].value_counts().head(1)

most_frequent('Rajasthan Royals')
```

```python
# 29. Find the team with the longest winning streak in a season.

streak = df.groupby(['Season','WinningTeam']).size().reset_index(name='Matches Won')
streak = streak[streak.groupby('Season')['Matches Won'].rank(method='first', ascending=False) == 1]
streak.reset_index(drop=True)
```

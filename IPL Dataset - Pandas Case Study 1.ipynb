```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ipl-matches.csv')
df
```

```python
# 1) Find all the final match winners.

df[df['MatchNumber'] == 'Final']
```

```python
# 2. How many Super Over finishes have occurred

sup_ov = df[df['SuperOver'] == 'Y']
len(sup_ov)
```

```python
# 3. How many matches has CSK won in Kolkata
df[(df['WinningTeam'] == 'Chennai Super Kings') & (df['City'] == 'Kolkata')].shape[0]
```

```python
# 4. Calculate the percentage of matches where the toss winner is also the match winner
(df[df['TossWinner'] == df['WinningTeam']].shape[0] / len(df)) * 100
```

```python
# 5. How many matches has each team played
t1 = df['Team1'].value_counts()
t2 = df['Team2'].value_counts()

t1.add(t2).sort_values(ascending=False)  

# fill_value=0 ensures that indices not present in one series are treated as 0 instead of NaN.
```

```python
# 6. Find the top 5 players who won the most Player of the Match awards
df['Player_of_Match'].value_counts().head()
```

```python
# 7. Find the player who won the most Player of the Match awards in Finals and Qualifiers

pom = df[(df['MatchNumber'] == 'Final') | (df['MatchNumber'] == 'Qualifier 1') | (df['MatchNumber'] == 'Qualifier 2') | (df['MatchNumber'] == 'Qualifier')]
pom['Player_of_Match'].value_counts().head(1)
```

```python
# 8. List all matches where a particular player played for both teams

# convert each cell from a string that looks like a list ("['A','B']") into an actual Python list (['A','B']).
import ast
df['Team1Players'] = df['Team1Players'].apply(ast.literal_eval)
df['Team2Players'] = df['Team2Players'].apply(ast.literal_eval)

# Solution
common_player = []
for i in range(len(df)):
    common = set(df.loc[i,'Team1Players']) & set(df.loc[i,'Team2Players'])
    common_player.append(bool(common))

df['common_players'] = common_player
df['common_players'].value_counts()  

# No player played for both teams
```

```python
# 9. Count how many matches each team has won in each city
df.groupby(['City','WinningTeam'])['ID'].count().reset_index(name='Wins')
```

```python
# 10. Find the highest winning margin by runs
df['Margin'].sort_values(ascending=False).head(1)
```

```python
# 11. Find the team with the most toss wins
df['TossWinner'].value_counts().head(1)
```

```python
# 12. Count how many matches were played in each season
df['Season'].value_counts()
```

```python
# 13. Find the team that won the most finals
fin_won = df[df['MatchNumber'] == 'Final']
fin_won['WinningTeam'].value_counts().head(1)
```

```python
# 14. Find the average margin of victory for each team
df.groupby(['WinningTeam'])['Margin'].mean()
```

```python
# 15. List matches where the toss winner lost the match
df[df['TossWinner'] != df['WinningTeam']]
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Simulated Missing Values Dataset.csv')
df1 = df.copy()
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Mean for numerical
imp_mean = SimpleImputer(strategy='mean')
df[['age']] = imp_mean.fit_transform(df[['age']])

# Median
imp_median = SimpleImputer(strategy='median')
df[['days_on_platform']] = imp_median.fit_transform(df[['days_on_platform']])

# Mode (most frequent) for categorical
#imp_mode = SimpleImputer(strategy='most_frequent')
#df[cat_cols] = imp_mode.fit_transform(df[cat_cols])

df.isnull().sum()
```

```python
# create df for grouped imputation
gr_df = df1.copy()
gr_df
```

```python
# Grouped mean example
gr_df['age'] = gr_df.groupby('gender')['age'].transform(
    lambda x: x.fillna(x.mean())
)

# Can replace x.mean() with x.median() or x.mode().iloc[0]

gr_df['days_on_platform'] = gr_df.groupby('city')['days_on_platform'].transform(
    lambda x: x.fillna(x.median())
)

gr_df.isnull().sum()
```

```python
# create df for constant value imputation
con_df = df1.copy()
con_df
```

```python
# Constant value for numerical
con_df['age'] = SimpleImputer(strategy='constant', fill_value=9999)
con_df.isnull().sum()

# Constant value for categorical
#con_df[col] = SimpleImputer(strategy='constant', fill_value='Missing')
```

```python
# new df
fb_df = df1.copy()

fb_df['age'] = fb_df['age'].ffill()
fb_df['days_on_platform'] = fb_df['days_on_platform'].bfill()

fb_df.isnull().sum()
```

```python
# create df

inter_df = df1.copy()

inter_df.infer_objects(copy=False)
inter_df.interpolate(method='linear', inplace=True)

# Other methods: 'time', 'polynomial', 'spline', etc.
#inter_df.interpolate(method='polynomial', order=2, inplace=True)
inter_df.isnull().sum()
```
